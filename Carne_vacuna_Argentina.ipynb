Esta es una manera mas sencilla para poder ver los valores unicos parte de excluir aquellos que no querramos ver

```python
# Lista de columnas a excluir
columnas_a_excluir = ['sector_id', 'variable_id', 'indice_tiempo','alcance_id','valor']

# Para obtener los valores únicos de cada columna (excluyendo las columnas especificadas):
for columna in dataframe.columns:
    if columna not in columnas_a_excluir:
        valores_unicos = dataframe[columna].unique()
        print(f"Valores únicos en la columna {columna}:\n{valores_unicos}\n")
```

```python
#Esta función devuelve un valor booleano para cada elemento del DataFrame, indicando si es True si el valor es nulo o False si no lo es.
dataframe.isnull()
```

```python
#Esta función es la inversa de isnull() y devuelve un valor booleano para cada elemento del DataFrame, indicando si es True si el valor no es nulo o False si lo es.
dataframe.notna()
```

```python
# indica la cantidad de valores no nulos (es decir, valores que no son NaN) en cada columna del DataFrame. En este caso, todas las columnas tienen 1221 valores no nulos, lo que significa que no hay valores nulos en ninguna de las columnas.
dataframe.notna().sum()
```
